## BERT Fine-Tuning

BERT (Bidirectional Encoder Representations from Transformers) ist ein Sprachmodell, das auf riesigen Textmengen vortrainiert wurde und dadurch bereits ein tiefes Verständnis von Sprache mitbringt. Beim Fine-Tuning wird dieses Vorwissen genutzt und das Modell zusätzlich auf einem spezifischen, annotierten Datensatz trainiert. Das Modell lernt dabei, welche sprachlichen Muster mit welcher Emotion zusammenhängen. Text wird dabei bidirektional verarbeitet, d.h. sowohl der vorherige als auch der nachfolgende Kontext jedes Wortes wird gleichzeitig berücksichtigt.

Fine-Tuning wird eingesetzt, wenn ein vortrainiertes Modell auf eine spezifische Aufgabe spezialisiert werden soll, ohne es von Grund auf neu zu trainieren. Es spart Rechenzeit und liefert in der Regel starke Ergebnisse, selbst bei begrenzten Trainingsdaten. Als Baseline wurde damit ein Macro-F1-Score von 0.49 erreicht.

Der F1-Score ist die zentrale Kennzahl zur Bewertung eines Klassifikationsmodells. Er kombiniert Precision (wie viele der als Joy erkannten Texte wirklich Joy waren) und Trefferquote (wie viele der echten Joy-Texte das Modell gefunden hat) zu einem einzigen Wert. Ein Modell das eine Klasse bevorzugt oder wahllos alles einer Klasse zuordnet, wird durch den F1-Score entsprechend bestraft. Der Wert liegt zwischen 0 und 1, wobei 1.0 eine perfekte Klassifikation bedeutet.

## Data Augmentation

Bei Emotionsdatensätzen ist es häufig der Fall, dass manche Emotionsklassen deutlich mehr Trainingsbeispiele haben als andere. Dieses Klassenungleichgewicht führt dazu, dass das Modell seltene Emotionen schlechter erkennt, was sich direkt negativ auf den F1-Score der betroffenen Klassen auswirkt. Data Augmentation begegnet diesem Problem, indem künstlich neue Trainingsbeispiele aus bestehenden Daten erzeugt werden.

Es gibt drei gängige Methoden zur Augmentation. Die erste Methode DDA ersetzt, vertauscht oder löscht Wörter zufällig in bestehenden Sätzen. Die zweite Methode nutzt BERT Embeddings, um Wörter durch semantisch ähnliche Alternativen zu ersetzen. Die dritte Methode ProtAugment formuliert ganze Sätze um, ohne die emotionale Bedeutung zu verändern. Die wirksamste Methode war ProtAugment, angewendet nur auf unterrepräsentierte Klassen, was den F1-Score von 0.49 auf 0.52 steigerte. Data Augmentation wird besonders dann eingesetzt, wenn der Datensatz unausgewogen ist und mehr Trainingsmaterial für bestimmte Klassen benötigt wird.

## Transfer Learning

Transfer Learning bezeichnet den Ansatz, ein Modell zunächst auf einem verwandten Datensatz zu trainieren und das dabei erworbene Wissen anschließend auf den eigentlichen Zieldatensatz zu übertragen. Es ermöglicht, Emotionssubtilitäten besser zu erfassen, ohne umfangreiche neue Trainingsdaten zu benötigen.

In der Praxis wird dazu zunächst ein Datensatz mit ähnlichen Emotionskategorien als Vortraining genutzt, bevor das Modell auf dem Zieldatensatz feinabgestimmt wird. Die Kombination aus Vortraining und ProtAugmentation erzielte den besten F1-Score von 0.52. Transfer Learning bietet sich an, wenn zwei thematisch verwandte Datensätze verfügbar sind, da das Vorwissen aus dem ersten Datensatz die Erkennungsleistung auf dem zweiten verbessert.

## RoBERTa

RoBERTa (Robustly Optimized BERT Pretraining Approach) ist eine Weiterentwicklung von BERT, die durch eine optimierte Trainingsstrategie robustere Sprachrepräsentationen erzeugt. Die Verbesserungen umfassen mehr Trainingsdaten, größere Batches und längere Trainingszeiten. Als Trainingsdaten dienen unter anderem die englische Wikipedia, BooksCorpus und weitere Web-Datensätze.

RoBERTa wird eingesetzt, wenn eine stärkere Baseline als BERT gefragt ist. Es ist jedoch zu beachten, dass ein neueres Modell nicht automatisch besser geeignet ist. Je nach Datensatz und Aufgabe kann BERT RoBERTa übertreffen, weshalb die Modellwahl immer aufgabenspezifisch getroffen werden sollte.

## Emotion-Aware RoBERTa

Emotion-Aware RoBERTa ist eine erweiterte Version von RoBERTa, die zwei zusätzliche Mechanismen integriert, um Schwächen bei der feingranularen Emotionserkennung zu beheben.

Der erste Mechanismus ist die Emotion-Specific Attention (ESA). Sie ist eine zusätzliche Schicht im Modell, die die Aufmerksamkeit dynamisch auf emotionsrelevante Wörter lenkt. Während die Standard-Attention alle Tokens gleichmäßig bewertet, lernt ESA durch einen trainierbaren Skalierungsvektor, welche Wörter für die Emotionsklassifikation besonders bedeutsam sind.

Der zweite Mechanismus ist das TF-IDF Gating. TF-IDF ist ein statistisches Maß, das die Bedeutsamkeit eines Wortes im Kontext eines Dokuments bewertet. Das Gating filtert informationsarme Tokens heraus und verstärkt die Gewichtung emotional bedeutsamer Inhalte. Besonders bei informellem Text mit Slang oder Abkürzungen verbessert dieser Mechanismus die Klassifikationsrobustheit erheblich.

Dieser Ansatz wird eingesetzt, wenn Standard-Transformer-Modelle an Grenzen stoßen, etwa bei Sarkasmus, impliziten Emotionsausdrücken oder stark unausgewogenen Klassen. Auf dem relevanten Kaggle-Datensatz mit sechs Emotionen wurde eine Genauigkeit von 96,77 % und ein F1-Score von 0.97 erreicht, was einer Verbesserung von 10,87 Prozentpunkten gegenüber dem Standard-RoBERTa entspricht.

## F1 - Formel

$$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$$
