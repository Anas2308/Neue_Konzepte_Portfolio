# Emotion Detection

## Emotion detection vs. Sentiment Analysis

Emotion detection and sentiment analysis are often used interchangeably but are differentiated by definition. [[1]](#alslaity-and-orij_2024)

> **Sentiment** is an attitude, thought or judgement prompted by a feeling

> **Emotion** is a concious mental reaction subjectively experienced as a strong feeling
