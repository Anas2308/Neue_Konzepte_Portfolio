```python
import polars as pl
```

## Load & Parse: /data/csv/tweet_emotion_dataset.csv

Source: https://www.kaggle.com/datasets/pashupatigupta/emotion-detection-from-text

```python
tweet_emotion = (
    pl.read_csv("../data/csv/tweet_emotion_dataset.csv")
    .rename({"sentiment": "emotion"})
    .select(["content", "emotion"])
    .with_row_index("tweet_id")
    .select(["tweet_id", "content", "emotion"])
)

tweet_emotion
```

## Load & Parse: /data/csv/emotion_sentiment_dataset.csv

Source: https://www.kaggle.com/datasets/simaanjali/emotion-analysis-based-on-text/data

```python
text_emotion = (
    pl.read_csv("../data/csv/emotion_sentiment_dataset.csv")
    .rename({"text" : "content", "Emotion" : "emotion"})
    .select(["content", "emotion"])
    .with_row_index("text_id")
    .select(["text_id", "content", "emotion"])
)

text_emotion
```

Test if both emotion categories align!

```python
tweet_emotion_categories = tweet_emotion.select("emotion").unique("emotion").sort("emotion")
emotion_sentiment_categories = text_emotion.select("emotion").unique("emotion").sort("emotion")

tweet_emotion_categories.equals(emotion_sentiment_categories)
```

# Export Parquets

```python
tweet_emotion.write_parquet("../data/tweet_emotion.parquet")

text_emotion.write_parquet("../data/text_emotion.parquet")
```
